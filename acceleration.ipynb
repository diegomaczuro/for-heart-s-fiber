```python
%%cython
import scipy
import scipy.io
import numpy as np

import cython
from cython.parallel import prange, parallel
import line_profiler

cdef int n = 25
cdef int slic = 100
filename = "./DTI080803/"

cdef _transpose(double[:,:] R, double[:,:] T):
    T[0,0] = R[0,0]
    T[0,1] = R[1,0]
    T[0,2] = R[2,0]
    
    T[1,0] = R[0,1]
    T[1,1] = R[1,1]
    T[1,2] = R[2,1]
    
    T[2,0] = R[0,2]
    T[2,1] = R[1,2]
    T[2,2] = R[2,2]

cdef double[:,:,:] in_log_euclidian(double[:,:,:] D, double[:,:,:] R):
    """
    This function transforms euclidian tensor into log_euclidian
    """
    
    cdef int i,j,k,c,d,k, summ
    
    cdef double[:,:] R_ = np.zeros((3,3))
    cdef double[:,:] R_T_= np.zeros((3,3))
    
    cdef double[:,:,:] log_D = np.zeros_like(D)
    cdef double[:,:,:] log_matrix = np.zeros_like(D)
    for i in prange(len(D), nogil=True):
        if D[i,0,0] < 0 or D[i,0,0] == 0:
            D[i,0,0] = 0.0001
        if D[i,1,1] < 0 or D[i,1,1] == 0:
            D[i,1,1] = 0.0001
        if D[i,2,2] < 0 or D[i,2,2] == 0:
            D[i,2,2] = 0.0001
        log_D[i,0,0] = np.log(D[i,0,0])
        log_D[i,1,1] = np.log(D[i,1,1])
        log_D[i,2,2] = np.log(D[i,2,2])
        
        return log_matrix

cdef read_tensor(str filename):
    
    cdef int koef = 100000
    cdef int i
    
    cdef double[:,:,:] e1 = scipy.io.loadmat(filename+'e1.mat')['e1']
    cdef double[:,:,:] e2 = scipy.io.loadmat(filename+'e2.mat')['e2']
    cdef double[:,:,:] e3 = scipy.io.loadmat(filename+'e3.mat')['e3']
    cdef double[:,:,:] v11 = scipy.io.loadmat(filename+'v11.mat')['v11']
    cdef double[:,:,:] v12 = scipy.io.loadmat(filename+'v12.mat')['v12']
    cdef double[:,:,:] v13 = scipy.io.loadmat(filename+'v13.mat')['v13']
    cdef double[:,:,:] v21 = scipy.io.loadmat(filename+'v21.mat')['v21']
    cdef double[:,:,:] v22 = scipy.io.loadmat(filename+'v22.mat')['v22']
    cdef double[:,:,:] v23 = scipy.io.loadmat(filename+'v23.mat')['v23']
    cdef double[:,:,:] v31 = scipy.io.loadmat(filename+'v31.mat')['v31']
    cdef double[:,:,:] v32 = scipy.io.loadmat(filename+'v32.mat')['v32']
    cdef double[:,:,:] v33 = scipy.io.loadmat(filename+'v33.mat')['v33']
    cdef double[:,:,:] D = np.zeros((n*n,3,3))
    cdef double[:,:,:] R = np.zeros((n*n,3,3))
    
    for i in xrange(n*n):
        D[i,0,0] = np.ravel(e1[0:n,0:n,slic])[i]
        D[i,0,0] = np.ravel(e1[0:n,0:n,slic])[i]
        D[i,0,0] = np.ravel(e1[0:n,0:n,slic])[i]*koef
        D[i,1,1] = np.ravel(e2[0:n,0:n,slic])[i]*koef
        D[i,2,2] = np.ravel(e3[0:n,0:n,slic])[i]*koef
        R[i,0,0] = np.ravel(v11[0:n,0:n,slic])[i]
        R[i,0,1] = np.ravel(v12[0:n,0:n,slic])[i]
        R[i,0,2] = np.ravel(v13[0:n,0:n,slic])[i]
        R[i,1,0] = np.ravel(v21[0:n,0:n,slic])[i]
        R[i,1,1] = np.ravel(v22[0:n,0:n,slic])[i]
        R[i,1,2] = np.ravel(v23[0:n,0:n,slic])[i]
        R[i,2,0] = np.ravel(v31[0:n,0:n,slic])[i]
        R[i,2,1] = np.ravel(v32[0:n,0:n,slic])[i]
        R[i,2,2] = np.ravel(v33[0:n,0:n,slic])[i]

    cdef double[:,:,:] log_matrix = in_log_euclidian(D,R)
    
    return log_matrix

def py_read_tensor(filename):
    return read_tensor(filename)

#read_tensor(filename)
#py_read_tensor(filename)
```

```python
%%cython
import numpy as np
from cython cimport view


cdef _transpose(double[:,:] R, double[:,:] T):
    T[0,0] = R[0,0]
    T[0,1] = R[1,0]
    T[0,2] = R[2,0]
    
    T[1,0] = R[0,1]
    T[1,1] = R[1,1]
    T[1,2] = R[2,1]
    
    T[2,0] = R[0,2]
    T[2,1] = R[1,2]
    T[2,2] = R[2,2]

cdef double[:,:,:] in_log_euclidian(double[:,:,:] D, double[:,:,:] R):
    """
    This function transforms euclidian tensor into log_euclidian
    """
    
    cdef int i,j,k
    
    cdef double[:,:] R_ = np.zeros((3,3))
    cdef double[:,:] R_T_= np.zeros((3,3))
    
    cdef double[:,:,:] log_D = np.zeros_like(D)
    cdef double[:,:,:] log_matrix = np.zeros_like(D)
    for i in xrange(len(D)):
        if D[i,0,0] < 0 or D[i,0,0] == 0:
            D[i,0,0] = 0.0001
        if D[i,1,1] < 0 or D[i,1,1] == 0:
            D[i,1,1] = 0.0001
        if D[i,2,2] < 0 or D[i,2,2] == 0:
            D[i,2,2] = 0.0001
        log_D[i,0,0] = np.log(D[i,0,0])
        log_D[i,1,1] = np.log(D[i,1,1])
        log_D[i,2,2] = np.log(D[i,2,2])
        
        R_[:, :] = R[i, :, :]
        _transpose(R_, R_T_)

        R_ = np.dot(np.dot(R[i,:,:],log_D[i,:,:]), R_T_)
        log_matrix[i] = R_
    return log_matrix

cdef double[:,:,:] D = np.zeros((1,3,3))
cdef double[:,:,:] R = np.zeros((1,3,3))

D[:,1,1] = 1
R[:,1,1] = 1

cdef double[:,:,:] res =  in_log_euclidian(D,R)
print np.array(res)
```
